```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.basketball-reference.com/players/j/jamesle01.html"
page = requests.get(URL)

html_doc = page.text
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
soup.body.find(id = 'wrap').find(id = 'meta').find('h1', {'itemprop': 'name'}).find('span').text
```

```python
soup.body.find(id = 'wrap').find(id = 'meta')
```

```python
soup.body.find(id = 'wrap').find(id = 'info').find(id = 'bling')
```

```python
soup.body.find(id = 'wrap').find(id = 'meta').find('img')['src']
```

```python
soup.body.find(id = 'wrap').find(id = 'info').find('div', {'class': 'uni_holder bbr'}).find('a', {'class': 'poptip'})['data-tip']
```

```python
teams = soup.body.find(id = 'wrap').find(id = 'info').find('div', {'class': 'uni_holder bbr'}).find_all('a', {'class': 'poptip'})
jersey = []

for team in teams:
  jersey.append(team['data-tip'])
```

```python
data = {
     'playerName': soup.body.find(id = 'wrap').find(id = 'meta').find('h1', {'itemprop': 'name'}).find('span').text, 
     'awards': soup.body.find(id = 'wrap').find(id = 'info').find(id = 'bling').text,
     'photoUrl': soup.body.find(id = 'wrap').find(id = 'meta').find('img')['src'],
     'teams': jersey
}

data
```
